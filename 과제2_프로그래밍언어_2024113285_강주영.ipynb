```python
#1번
#  ref-덧셈 곱셈 수식기 파서(2024.09.25).docx -  
# 3주차에 제공한 파서 샘플 프로그램을 실행 


class Parser:
    def __init__(self, expression):
        self.tokens = list(expression.replace(" ", ""))  # 공백을 제거하고 수식을 토큰화
        self.current_token = None  # 현재 토큰
        self.next_token()  # 첫 번째 토큰으로 이동

    def next_token(self):
        """다음 토큰으로 이동"""
        if len(self.tokens) > 0:
            self.current_token = self.tokens.pop(0)  # 리스트의 첫 번째 토큰을 가져옴
        else:
            self.current_token = None  # 더 이상 토큰이 없으면 None으로 설정

    def match(self, token_type):
        """현재 토큰이 기대한 토큰과 일치하는지 확인하고 일치하면 소비"""
        if self.current_token == token_type:
            self.next_token()  # 토큰을 소비하고 다음 토큰으로 이동
        else:
            raise Exception(f"Unexpected token: {self.current_token}")

    def parse(self):
        """수식 분석 시작"""
        return self.expr()

    def expr(self):
        """expr -> term { + term }"""
        result = self.term()  # 첫 번째 term 계산
        while self.current_token == '+':  # 덧셈이 계속 나오면 반복
            self.match('+')  # '+' 기호 소비
            result += self.term()  # 다음 term을 계산하고 더함
        return result

    def term(self):
        """term -> factor { * factor }"""
        result = self.factor()  # 첫 번째 factor 계산
        while self.current_token == '*':  # 곱셈이 계속 나오면 반복
            self.match('*')  # '*' 기호 소비
            result *= self.factor()  # 다음 factor를 계산하고 곱함
        return result

    def factor(self):
        """factor -> number | ( expr )"""
        if self.current_token == '(':
            self.match('(')  # '(' 괄호를 소비
            result = self.expr()  # 괄호 안의 수식을 처리
            self.match(')')  # ')' 괄호를 소비
            return result
        else:
            return self.number()  # 숫자를 처리

    def number(self):
        """여러 자릿수 숫자 처리"""
        result = 0
        while self.current_token is not None and self.current_token.isdigit():
            result = result * 10 + int(self.current_token)
            self.next_token()  # 숫자를 모두 처리
        return result

# 사용자 입력 수식을 파싱하고 결과 출력
def calculate(expression):
    parser = Parser(expression)
    result = parser.parse()
    return result

# 예시 실행
expression = "2+(3+4) * 5"
result = calculate(expression)
print(f"Result: {result}")
```

```python
#2번
#  파서 프로그램에서 2+3*(4+5)*6 를 실행 시 순차적으로 진행 과정을 요약 설명

class Parser:
    def __init__(self, expression): 
        #expression에 2+3*(4+5)*6 
        self.tokens = list(expression.replace(" ", ""))  
        #expression의 값에 공백이 있다면 제거하고 해당 문자열을 리스트 형식으로 변경 & 
        self.current_token = None  
        # 현재 토큰에 값의 부재라는 것을 나타냄(None)
        self.next_token()  
        # next_token 함수 호출
    
    def next_token(self):
        """다음 토큰으로 이동"""
        if len(self.tokens) > 0:
            # self.tokens의 길이가 0보다 크다면 
            # len() 해당 값의 길이 반환
            self.current_token = self.tokens.pop(0)  
            #self.current_token에 self.tokens의 첫 번째 값(토큰)을 가져와 넣어줌
        else:
            self.current_token = None  
            #self.tokens의 값이 0보다 작거나 같으면 현재토큰에 None 설정
            #self.tokens.pop()을 하는 것 자체가 리스트에 값을 단순 출력하는 것이 아니라, 리스트의 값을 아예 추출하는 것.
    
    def match(self, token_type):
        """현재 토큰이 기대한 토큰과 일치하는지 확인하고 일치하면 소비"""
        if self.current_token == token_type:
            self.next_token()  # 토큰을 소비하고 다음 토큰으로 이동
        else:
            raise Exception(f"Unexpected token: {self.current_token}")
            # 현재 토큰이 기대하는 토큰과 일치하지 않으면 예외를 보냅니다. 
            # raise : try-exception을 사용하지 않고, 원하는 결과가 아닐 때 예외를 발생시킬 수 있다.

    def parse(self):
        """수식 분석 시작"""
        return self.expr() 
        # self.expr()호출하여 반환 된 값을 반환

    def expr(self):
        """expr -> term { + term }"""
        result = self.term() 
        #term함수 호출. 첫번째 항을 계산한다. 
        
        while self.current_token == '+':  
            # 현재 토큰이 '+' 인 경우 반복
            self.match('+')  
            # self.match('+') 호출하여 처리(소비)
            result += self.term() 
            # 다음 항을 계산하여 결과에 더하기

        return result
        # 결과 반환
    
    def term(self):
        """term -> factor { * factor }"""
        result = self.factor()  

        #factor 함수 호출
        # 첫 번째 인자(숫자)를 계산
        
        while self.current_token == '*':  
            #현재 토큰이 '*'인 동안 반복
            self.match('*')  
            # self.match('*') 호출하여 처리(소비)
            result *= self.factor()  
            # 다음 인자(숫자)를 계산하여 결과에 곱합
        return result
        #계산된 결과 반환

    def factor(self):
        """factor -> number | ( expr )"""
        if self.current_token == '(':
            #만약 현재토큰 값이 '(' 라면
            self.match('(')  
            # '(' 괄호를 소비
            # self.match('(') 호출 > ( 괄호 소비. 
            result = self.expr()  
            # 괄호 안의 표현식을 분석후 반환 값 결과(result에 넣기)
            self.match(')')
            # ')' 괄호를 소비
            # self.match(')') 호출 > ) 괄호 소비.     
            return result
            #결과 반환
        else:
            return self.number()  
            #현재 토큰 값이 괄호가 아니라면 숫자를 반환

    def number(self):
        """여러 자릿수 숫자 처리"""
        result = 0
        #result를 0으로 설정 
        while self.current_token is not None and self.current_token.isdigit():
            # 현재 토큰이 None이 아니고, isdigit()의 반환값이 숫자인 경우에 반복
            # -> 현재 토큰이 숫자일 동안 반복
            result = result * 10 + int(self.current_token)
            #이전 결과(result) 10을 곱하고 현재 토큰값을 더하기
            #현재 토큰값을 int(정수)로 형변환
            self.next_token() 
            #다음 토큰으로 이동하여 숫자를 모두 처리
        return result
        #최종적으로 계산된 숫자를 반복

# 사용자 입력 수식을 파싱하고 결과 출력
#expression : 2+3*(4+5)*6
def calculate(expression): 
    parser = Parser(expression) 
    # Parser클래스의 인스턴스를 생성
    # Parser 클래스의 __init__ (생성자) 호출 
    result = parser.parse()
    #Parser 클래스의 parser 메서드 호출
    #수식을 분석하여 결과를 얻기
    return result
    # 분석된 수식 반환

# 예시 실행
expression = "2+3*(4+5)*6"
result = calculate(expression) 
#calculate 함수 호출 >> 매개변수에 "2+3*(4+5)*6" 들어감
print(f"Result: {result}")
#반환된 값 출력력
```

```python
#2-3번
# -  Expr, term 함수 실행 수  print 문으로 인쇄 


class Parser:
    def __init__(self, expression):
        self.tokens = list(expression.replace(" ", ""))  
        self.current_token = None  
        self.next_token()  
        self.expr_count = 0  # expr 호출 횟수
        self.term_count = 0  # term 호출 횟수

    def next_token(self):
        """다음 토큰으로 이동"""
        if len(self.tokens) > 0:
            self.current_token = self.tokens.pop(0)
        else:
            self.current_token = None  

    def match(self, token_type):
        """현재 토큰이 기대한 토큰과 일치하는지 확인하고 일치하면 소비"""
        if self.current_token == token_type:
            self.next_token() 
        else:
            raise Exception(f"Unexpected token: {self.current_token}")

    def parse(self):
        """수식 분석 시작"""
        return self.expr()

    def expr(self):
        """expr -> term { + term }"""
        self.expr_count += 1  # expr 호출 횟수 증가
        print(f"expr 호출 횟수: {self.expr_count}")  # expr 함수 호출 횟수 출력
        result = self.term()  
        while self.current_token == '+':  
            self.match('+')  
            result += self.term()
        return result

    def term(self):
        """term -> factor { * factor }"""
        self.term_count += 1  # term 호출 횟수 증가
        print(f"term 호출 횟수: {self.term_count}")  # term 함수 호출 횟수 출력
        result = self.factor() 
        while self.current_token == '*':  
            self.match('*')  
            result *= self.factor()  
        return result

    def factor(self):
        """factor -> number | ( expr )"""
        if self.current_token == '(':
            self.match('(')  
            result = self.expr()  
            self.match(')')  
            return result
        else:
            return self.number()  

    def number(self):
        """여러 자릿수 숫자 처리"""
        result = 0
        while self.current_token is not None and self.current_token.isdigit():
            result = result * 10 + int(self.current_token)
            self.next_token() 
        return result

# 사용자 입력 수식을 파싱하고 결과 출력
def calculate(expression):
    parser = Parser(expression)
    result={"result" : parser.parse(), "expr_count": parser.expr_count, "term_count": parser.term_count}
    
    return result
    

expression = "2+3*(4+5)*6"
result = calculate(expression)

print("---------")
print(f"Result: {result}")
for key, value in result.items():
    print(key, ":",  value)
print("---------")
```
